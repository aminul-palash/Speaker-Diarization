```
from datasets import load_dataset

HF_DATASET = "aminulpalash506/voxconverse-30-subset"

def inspect_dataset():
    dataset = load_dataset(HF_DATASET, split="train")

    print("\n================ DATASET INFO ================\n")

    # 1. Column names
    print("📌 Columns:")
    print(dataset.column_names)

    # 2. First sample raw structure
    print("\n📌 First sample keys:")
    sample = dataset[0]
    print(sample.keys())

    print("\n📌 Full first sample:")
    for k, v in sample.items():
        print(f"\n--- {k} ---")
        print(type(v))
        print(v if len(str(v)) < 500 else str(v)[:500] + "...")

    # 3. Check likely GT fields
    print("\n📌 Ground-truth candidates check:")

    possible_keys = ["segments", "annotation", "rttm", "speakers", "labels"]

    for k in possible_keys:
        print(f"{k}: {'YES' if k in sample else 'NO'}")

    # 4. Audio format check
    print("\n📌 Audio format:")
    audio = sample.get("audio", None)
    if audio:
        print(audio.keys())
        print("sample_rate:", audio.get("sampling_rate", None))

    print("\n============================================\n")

if __name__ == "__main__":
    inspect_dataset()
```
